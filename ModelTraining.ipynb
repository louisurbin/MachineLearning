Equation normale

```python
import numpy as np
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100,1)

X_b = np.c_[np.ones((100, 1)), X]  # On ajoute x0=1 à chaque observation
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
theta_best
```

```python
# Prédictions
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new]
y_predict = X_new_b.dot(theta_best)
y_predict
```

```python
import matplotlib  # On affiche les prédictions
import matplotlib.pyplot as plt
plt.plot(X_new, y_predict, "r-")
plt.plot(X, y, "b.")
plt.axis([0, 2, 0, 15])
plt.show()
```

```python
# Equivalence en Scikit-Learn
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
lin_reg.intercept_, lin_reg.coef_
```

```python
lin_reg.predict(X_new)
```

Descente de gradient

```python
# Descente de gradient ordinaire
n_iterations = 10
m = 100
fig, axs = plt.subplots(1, 3, figsize=(18, 5))
for idx, eta in enumerate([0.02, 0.1, 0.5]):  # Taux d'apprentissage
    theta = np.random.randn(2, 1)  # Initialisation
    for i in range(n_iterations):
        gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y)
        theta = theta - eta * gradients
        y_predict = X_new_b.dot(theta)
        axs[idx].plot(X_new, y_predict, "r-")
    axs[idx].plot(X, y, "b.")
    axs[idx].axis([0, 2, 0, 15])
    axs[idx].set_title(f"$\eta$ = {eta}")
plt.show()
```

```python
# A gauche, le taux d'apprentissage est trop faible, à droite il est trop fort et au milieu il est bon
```

```python
# Descente de gradient stochastique
n_epochs = 10
t0, t1 = 5, 50
m = 1

def learning_schedule(t):
    return t0 / (t + t1)

theta = np.random.randn(2,1)  # Initialisation aléatoire

for epoch in range(n_epochs):
    for i in range(m):
        random_index = np.random.randint(m)
        xi = X_b[random_index:random_index+1]
        yi = y[random_index:random_index+1]
        gradients = 2 * xi.T.dot(xi.dot(theta) - yi)
        eta = learning_schedule(epoch * m + i)
        theta = theta - eta * gradients
    y_predict = X_new_b.dot(theta)
    plt.plot(X_new, y_predict, "r-")

plt.plot(X, y, "b.")
plt.axis([0, 2, 0, 15])
plt.show()
```

```python
# Equivalence en Scikit-Learn
from sklearn.linear_model import SGDRegressor
sgd_reg = SGDRegressor(max_iter=50, penalty=None, eta0=0.1)
sgd_reg.fit(X, y.ravel())
sgd_reg.intercept_, sgd_reg.coef_
```

```python
# Autre idée : descente de gradient par mini-lots
```

Courbes d'apprentissage

```python
# Fonction qui trace les courbes d'apprentissage d'un modèle pour un jeu d'entraînement donné.
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

def plot_learning_curves(model, X, Y):
    X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2)
    train_errors, val_errors = [], []
    for m in range (1, len(X_train)):
        model.fit(X_train[:m], y_train[:m])
        y_train_predict = model.predict(X_train[:m])
        y_val_predict = model.predict(X_val)
        train_errors.append(mean_squared_error(y_train_predict, y_train[:m]))
        val_errors.append(mean_squared_error(y_val_predict, y_val))
    plt.plot(np.sqrt(train_errors), "r-+", linewidth=2, label="Jeu d'entraînement")
    plt.plot(np.sqrt(val_errors), "b-", linewidth=3, label="Jeu de validation")
    plt.legend()
    plt.xlabel("Taille du jeu d'entraînement")
    plt.ylabel("RMSE")
    plt.axis([0, 80, 0, 3])
    plt.show()
```

```python
# Modèle linéaire
lin_reg = LinearRegression()
plot_learning_curves(lin_reg, X, y)
```

```python
# Modèle polynomial d'ordre 10
from sklearn.pipeline import Pipeline
polynomial_regression = Pipeline((
        ("poly_features", PolynomialFeatures(degree=10, include_bias=False)),
        ("lin_reg", LinearRegression()),
    ))
plot_learning_curves(polynomial_regression, X, y)
```

```python
# Ecart entre les 2 courbes car modèle qui surajuste
```

```python
# Arrêt précoce
from sklearn.base import clone
from sklearn.preprocessing import StandardScaler
import joblib

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2)
y_train = np.ravel(y_train)
y_val = np.ravel(y_val)

# Normalisation
scaler = StandardScaler()
X_train_poly_scaled = scaler.fit_transform(X_train)
X_val_poly_scaled = scaler.transform(X_val)

sgd_reg = SGDRegressor(max_iter=100, warm_start=True, penalty=None, learning_rate="constant", eta0=0.0005)
# warm_start=True signifie que la méthode fit, lorsqu'elle est appelée reprend l'entraînement là où il a été
# interrompu au lieu de recommencer entièrement.

minimum_val_error = float("inf")
best_epoch = None
best_model = None
for epoch in range(1000):
    sgd_reg.fit(X_train_poly_scaled, y_train)
    y_val_predict = sgd_reg.predict(X_val_poly_scaled)
    val_error = mean_squared_error(y_val_predict, y_val)
    if val_error < minimum_val_error:
        minimum_val_error = val_error
        best_epoch = epoch
        best_model = clone(sgd_reg)
joblib.dump(best_model, 'best_model.pkl')

print(f"Best epoch: {best_epoch}")
print(f"Minimum validation error: {minimum_val_error}")
print("Best model saved as 'best_model.pkl'")
```

Régression logistique

```python
from sklearn import datasets
iris = datasets.load_iris()
list(iris.keys())
```

```python
X = iris["data"][:, 3:]
y = (iris["target"] == 2).astype(int)  # 1 si iris virginica, 0 sinon

from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X,y)

X_new = np.linspace(0, 3, 1000).reshape(-1, 1)
y_proba = log_reg.predict_proba(X_new)

plt.plot(X_new, y_proba[:, 1], "g-", label="Iris virginica")
plt.plot(X_new, y_proba[:, 0], "b--", label="Iris non-virginica")
plt.xlabel("Largeur de pétale(cm)")
plt.ylabel("Probabilité")
plt.legend()
plt.show()
```

```python
log_reg.predict([[1.67], [1.66]])
```

Chargement des données

```python
from sklearn.datasets import fetch_openml  # Chargement de la base de données MNIST originale
mnist = fetch_openml('mnist_784', version=1, parser='auto')
```

```python
X, y = mnist['data'], mnist['target']  # X et y sont des DataFrame et non des tableaux NumPy
X = X.to_numpy()
y = y.to_numpy()
y = y.astype(int)  # Important sinon les éléments de y ne sont pas considérés comme des entiers
X.shape
```

```python
y.shape
```

```python
import matplotlib  # On affiche un chiffre à partir d'un éléments de la base de donnée
import matplotlib.pyplot as plt
some_digit = X[5316]
some_digit_image = some_digit.reshape(28,28)
plt.imshow(some_digit_image, cmap=matplotlib.cm.binary, interpolation="nearest")
plt.axis("off")
plt.show()
```

```python
y[5316]
```

```python
X_train, X_test, y_train, y_test = X[:60000], X[60000:], y[:60000], y[60000:]
```

```python
import numpy as np 
shuffle_index = np.random.permutation(60000)
X_train, y_train = X_train[shuffle_index], y_train[shuffle_index]
```

Classificateur binaire

```python
y_train_9 = (y_train == 9)  # Vrai pour les 9, faux sinon
y_test_9 = (y_test == 9)
```

```python
from sklearn.linear_model import SGDClassifier  # Classificateur de descente de gradient stochastique
sgd_clf = SGDClassifier(random_state=42)
sgd_clf.fit(X_train, y_train_9)
```

```python
# Test sur some_digit
sgd_clf.predict([some_digit])  # Trop exotique
```

Mesures de performances

```python
# Implémentation d'une validation croisée
from sklearn.model_selection import StratifiedKFold
from sklearn.base import clone 

skfolds = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)

for train_index, test_index in skfolds.split(X_train, y_train_9):
    clone_clf = clone(sgd_clf)  # Utilisation d'un modèle non entraîné à chaque itération
    X_train_folds = X_train[train_index]
    y_train_folds = y_train_9[train_index]
    X_test_fold = X_train[test_index]
    y_test_fold = y_train_9[test_index]

    clone_clf.fit(X_train_folds, y_train_folds)
    y_pred = clone_clf.predict(X_test_fold)
    n_correct = sum(y_pred == y_test_fold)
    print(n_correct / len(y_pred))  # Pourcentages de prédictions correctes pour les 3 coupes
```

```python
# Validation croisée avec cross_val_score()
from sklearn.model_selection import cross_val_score
cross_val_score(sgd_clf, X_train, y_train_9, cv=3, scoring="accuracy")
```

```python
# Pas si impressionant que ça!
from sklearn.base import BaseEstimator
class Never9Classifier(BaseEstimator):  # classificateur qui se contente de classer dans la classe "non 9"
    def fit(self, X, y=None):
        pass
    def predict(self, X):
        return np.zeros((len(X),1), dtype=bool)
```

```python
never_9_clf = Never9Classifier()
cross_val_score(never_9_clf, X_train, y_train_9, cv=3, scoring="accuracy")  # En effet les 9 ne représentent qu'environ 10% des images
```

```python
# L'exactitude n'est pas en général la mesure de performance de prédilection pour les classificateurs, 
# surtout si certaines classes sont bien plus fréquentes que d'autres (jeux de données asymétriques)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_train_9, y_train_pred)  # ligne : classe réelle, colonne : classe prédite
```

```python
# Métrique plus concise : Précision et Rappel (Recall)
```

```python
recall_score(y_train_9, y_train_pred)
```

```python
# Comment décider du seuil à utiliser ?
y_scores = cross_val_predict(sgd_clf, X_train, y_train_9, cv=3, method="decision_function")
```

```python
from sklearn.metrics import precision_recall_curve
precisions, recalls, thresholds = precision_recall_curve(y_train_9, y_scores)
```

```python
def plot_precision_recall_vs_threshold(precisions, recalls, thresholds):
    plt.plot(thresholds, precisions[:-1], "b--", label="Précision")
    plt.plot(thresholds, recalls[:-1], "g-", label="Rappel")
    plt.xlabel("Seuil")
    plt.legend(loc="upper left")
    plt.ylim([0,1])
               
plot_precision_recall_vs_threshold(precisions, recalls, thresholds)  # Courbe de précision plus irrégulière
plt.show()
```

```python
plt.plot(recalls, precisions, "r-")
plt.xlabel("Rappel")
plt.ylabel("Précision")
plt.show()
```

```python
# Pour une précision de 90%, on choisit threshold = 12500
threshold = 12500
y_train_pred_90 = (y_scores > threshold)
precision_score(y_train_9, y_train_pred_90)
```

```python
recall_score(y_train_9, y_train_pred_90)  # Le rappel est bas
```

```python
# Courbe ROC : fpr=sensibilté en fonction de tpr=1-spécificité
from sklearn.metrics import roc_curve
fpr, tpr, thresholds = roc_curve(y_train_9, y_scores)
```

```python
def plot_roc_curve(fpr, tpr, label=None):
    plt.plot(fpr, tpr, linewidth=2, label=label)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.axis([0, 1, 0, 1])
    plt.xlabel('Taux de faux positifs')
    plt.ylabel('Taux de vrais positifs')

plot_roc_curve(fpr, tpr)
plt.show()
```

```python
# Remarque : Préférez la courbe PR si la classe positive est rare ou si vous attachez d'avantage 
# d'importance aux faux positifs qu'aux faux négatifs, et la courbe ROC dans le cas contraire
```

```python
from sklearn.ensemble import RandomForestClassifier  # A présent on entraîne une forêt aléatoire
forest_clf = RandomForestClassifier(random_state=42)
y_probas_forest = cross_val_predict(forest_clf, X_train, y_train_9, cv=3, method="predict_proba")
y_scores_forest = y_probas_forest[:, 1]  # Probabilités prédites pour la classe 9
fpr_forest, tpr_forest, thresholds_forest = roc_curve(y_train_9, y_scores_forest)
```

```python
# Plot des 2 courbes ROC (SGD et Forêt aléatoire)
plt.plot(fpr, tpr, "b:", label="SGD")
plot_roc_curve(fpr_forest, tpr_forest, "Forêt aléatoire")
plt.legend(loc="lower right")
plt.show()
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_train_9, y_scores_forest)  # Aire sous la courbe ROC
```

```python
precisions_forest, recalls_forest, thresholds_forest = precision_recall_curve(y_train_9, y_scores_forest)
plot_precision_recall_vs_threshold(precisions_forest, recalls_forest, thresholds_forest)
plt.show()
```

```python
plt.plot(recalls_forest, precisions_forest, "r-")
plt.xlabel("Rappel")
plt.ylabel("Précision")
plt.show()
```

```python
y_train_pred_forest = cross_val_predict(forest_clf, X_train, y_train_9, cv=3)
print("Précision pour la forêt aléatoire : ", precision_score(y_train_9, y_train_pred_forest))
```

```python
print("Rappel pour la forêt aléatoire : ", recall_score(y_train_9, y_train_pred_forest))
```

Analyse des erreurs

```python
y_train_pred = cross_val_predict(sgd_clf, X_train_scaled, y_train, cv=3)
conf_mx = confusion_matrix(y_train, y_train_pred)
conf_mx  # Représentation peu pratique
```

```python
plt.matshow(conf_mx, cmap=plt.cm.gray)
plt.show()  # Matrice diagonale : images correctement identifiées dans l'ensemble
```

```python
# On peut afficher les erreurs 
row_sums = conf_mx.sum(axis=1, keepdims=True)
norm_conf_mx = conf_mx / row_sums  # Taux d'erreur
np.fill_diagonal(norm_conf_mx, 0)  # Diagonale de 0 pour ne conserver que les erreurs
plt.matshow(norm_conf_mx, cmap=plt.cm.gray)
plt.show()
```

```python
# Fonction plot_digits
def plot_digits(instances, images_per_row, **options):
    size = 28
    images_per_row = min(len(instances), images_per_row)
    images = [instance.reshape(size, size) for instance in instances]
    n_rows = (len(instances) - 1) // images_per_row + 1
    row_images = []
    n_empty = n_rows * images_per_row - len(instances)
    images.append(np.zeros((size, size * n_empty)))
    for row in range(n_rows):
        rimages = images[row * images_per_row: (row + 1) * images_per_row]
        row_images.append(np.concatenate(rimages, axis=1))
    image = np.concatenate(row_images, axis=0)
    plt.imshow(image, cmap="binary", **options)
    plt.axis("off")
```

```python
# Echantillons de 3 et de 5
cl_a, cl_b = 3, 5
X_aa = X_train[(y_train == cl_a) & (y_train_pred == cl_a)]
X_ab = X_train[(y_train == cl_a) & (y_train_pred == cl_b)]
X_ba = X_train[(y_train == cl_b) & (y_train_pred == cl_a)]
X_bb = X_train[(y_train == cl_b) & (y_train_pred == cl_b)]
plt.figure(figsize=(8,8))
plt.subplot(221)
plot_digits(X_aa[:25], images_per_row=5)
plt.subplot(222)
plot_digits(X_ab[:25], images_per_row=5)
plt.subplot(223)
plot_digits(X_ba[:25], images_per_row=5)
plt.subplot(224)
plot_digits(X_bb[:25], images_per_row=5)
plt.show()
```

```python
# La plupart des images mal classées sont nous paraissent être des erreurs évidentes.
# Explication : SGDClassifier est un modèle linéaire qui ne fait qu'attribuer un poids par classe à chaque pixel.
# Modèle sensible aux translations/rotations, nécessité d'un prétraitement pour centrer et orienter les images
```

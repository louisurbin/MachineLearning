Classification SVM linéaire

```python
# Entraînement modèle SVM avec LinearSVC
import numpy as np 
from sklearn import datasets
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC

iris = datasets.load_iris()
X = iris["data"][:, (2, 3)]  # On charge les longeurs et largeurs des pétales
y = (iris["target"] == 2).astype(np.float64)

svm_clf = Pipeline((
        ("scaler", StandardScaler()),
        ("linear_svc", LinearSVC(C=1, loss="hinge", dual=True)),
    ))
svm_clf.fit(X, y)
```

```python
svm_clf.predict([[5.5, 1.7]])
```

Classification SVM non linéaire

```python
from sklearn.datasets import make_moons
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures

X, y = make_moons(n_samples=100, noise=0.15, random_state=42)

polynomial_svm_clf = Pipeline((
        ("poly_features", PolynomialFeatures(degree=3)),
        ("scaler", StandardScaler()),
        ("svm_clf", LinearSVC(C=10, loss="hinge", dual=True, max_iter=10000)),
    ))
polynomial_svm_clf.fit(X, y)
```

```python
import matplotlib.pyplot as plt
# Fonction pour plot la frontière
def plot_decision_boundary(clf, X, y, axes):
    x0s = np.linspace(axes[0], axes[1], 100)
    x1s = np.linspace(axes[2], axes[3], 100)
    x0, x1 = np.meshgrid(x0s, x1s)
    X_new = np.c_[x0.ravel(), x1.ravel()]
    y_pred = clf.predict(X_new).reshape(x0.shape)
    plt.contourf(x0, x1, y_pred, alpha=0.3)
    plt.scatter(X[:, 0], X[:, 1], c=y, s=30, edgecolor="k")
    plt.axis(axes)

plt.figure(figsize=(10, 6))
plot_decision_boundary(polynomial_svm_clf, X, y, axes=[-1.5, 2.5, -1, 1.5])
plt.show()
```

```python
# Noyau polynomial
from sklearn.svm import SVC
poly_kernel_svm_clf = Pipeline((
        ("scaler", StandardScaler()),
        ("svm_clf", SVC(kernel="poly", degree=10, coef0=100, C=5)),
    ))
poly_kernel_svm_clf.fit(X, y)
plt.figure(figsize=(10, 6))
plot_decision_boundary(poly_kernel_svm_clf, X, y, axes=[-1.5, 2.5, -1, 1.5])
plt.show()
```

```python
# Noyau radial gaussien
rbf_kernel_svm_clf = Pipeline((
        ("scaler", StandardScaler()),
        ("svm_clf", SVC(kernel="rbf", gamma=5, C=0.001))
    ))
rbf_kernel_svm_clf.fit(X, y)

plt.figure(figsize=(10, 6))
plot_decision_boundary(rbf_kernel_svm_clf, X, y, axes=[-1.5, 2.5, -1, 1.5])
plt.show()
```

Régression SVM

```python
from sklearn.svm import LinearSVR

X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100,1)
y = y.ravel()

svm_reg = Pipeline((
        ("scaler", StandardScaler()),
        ("svm_reg", LinearSVR(epsilon=1.5, dual=True)),
    ))
svm_reg.fit(X, y)
y_pred = svm_reg.predict(X)

plt.plot(X, y, "b.")
plt.plot(X, y_pred, "r-", label="régression")
plt.title("$\epsilon = 1.5$")
plt.legend()
plt.show()
```

```python
from sklearn.svm import SVR

X = 2 * np.random.rand(100, 1)
y = X**2 - 2 * X + 1 + 0.2 * np.random.randn(100,1)

sorted_indices = np.argsort(X[:, 0])
X = X[sorted_indices]
y = y[sorted_indices]

y = y.ravel()

svm_poly_reg = Pipeline((
        ("scaler", StandardScaler()),
        ("svm_reg", SVR(kernel="poly", degree=2, C=100, epsilon=0.1)),
    ))
svm_poly_reg.fit(X, y)
y_pred = svm_poly_reg.predict(X)

sorted_indices = np.argsort(X[:, 0])
X_sorted = X[sorted_indices]
y_sorted = y[sorted_indices]

plt.plot(X, y, "b.")
plt.plot(X, y_pred, "r-", label="régression")
plt.title("degré = 2, C=100, $\epsilon = 0.2$")
plt.legend()
plt.show()
```
